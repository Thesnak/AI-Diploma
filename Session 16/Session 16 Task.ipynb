```python
#matplotlib 
#seaborn 
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv("nobel_prize_by_winner.csv")
```

```python
allcountries = data.country.unique()

allcounts = []
for i in allcountries:
    allcounts.append(data.loc[data.country == i].count()[0])
    
plt.pie(allcounts, labels=allcountries, radius=1)
plt.show()
```

```python
%matplotlib notebook
plt.style.use('seaborn-notebook')

maxcountries = data.country.value_counts()[:8].to_list()
maxcountriesnames = data.country.value_counts().index[:8] 

maxcountriesnames = maxcountriesnames.to_list()
maxcountriesnames.append("Others")
allcountries = sum(maxcountries)

others = sum(list(data.country.value_counts())) - allcountries
maxcountries.append(others)

plt.bar(maxcountriesnames,maxcountries,)
plt.show()
```

```python
d=data[['category','country']]
```

```python
d=d.dropna()
```

```python
allcategory = d.category.unique()
allcountries =d.country.unique()
```

```python
len(d[(d.category=='economics')&(d.country=='USA')])
```

```python
allcategory=allcategory.tolist()
```

```python
dicCatergoryCountries={}
for country in allcountries :
    l=[]
    for category in allcategory :
        l.append(len(d[(d.country==country)&(d.category==category)]))
    dicCatergoryCountries[country]=l
```

```python
dicCatergoryCountries
```

```python
plt.pie([1, 0, 1, 0, 0, 9, 0],labels=allcategory)
plt.show()
```

```python
i=1
for k,v in dicCatergoryCountries.items():
    plt.subplot(2,2,i)
    plt.title(k)
    plt.pie(v,labels=allcategory)
    i+=1
    if i == 4:
        break
plt.show() 
```
